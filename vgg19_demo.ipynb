# VGG19 Exploration

```python
import torch
import torchvision.models as models
import torch.nn as nn
```

```python
vgg19 = models.vgg19(weights=models.VGG19_Weights.DEFAULT)
```

```python
print(vgg19)
```

```python
from torchsummary import summary

vgg19.to('cuda' if torch.cuda.is_available() else 'cpu')
summary(vgg19, input_size=(3, 224, 224))
```

```python
# Freeze feature layers
for param in vgg19.features.parameters():
    param.requires_grad = False
```

```python
vgg19.to('cuda' if torch.cuda.is_available() else 'cpu')
summary(vgg19, input_size=(3, 224, 224))
```

```python
vgg19.classifier[6] = nn.Linear(4096, 10)
```

```python
vgg19.to('cuda' if torch.cuda.is_available() else 'cpu')
summary(vgg19, input_size=(3, 224, 224))
```

```python
for name, layer in vgg19.features.named_children():
    print(f"Layer name: {name}, Layer: {layer}")
```

```python
for idx, layer in enumerate(vgg19.features):
    if isinstance(layer, torch.nn.Conv2d):
        print(f"Conv2D Layer index: {idx}, Layer: {layer}")
```

```python
for idx, layer in enumerate(vgg19.features):
    if isinstance(layer, torch.nn.MaxPool2d):
        print(f"Conv2D Layer index: {idx}, Layer: {layer}")
```

```python
for name, layer in vgg19.named_modules():
    print(layer)
```
