# Pandas Demo

```python
#importing required packages

import pandas as pd
```

## Dataframes

- Main data structure used by Pandas.
- Can use any sort of tabular data.
- Provides a bunch of functionalities on top to this tabular data.

```python
# Load dataset into a dataframe

df = pd.read_csv("datasets/diabetes.csv")

# TO do other file formats, do
# read_parquet (BEST)
# read_excel
# read_feather

# To change file formats
# pd.to_csv
# pd.to_parquet
# pd.to_excel
# pd.to_feather
```

```python
# Preview the first few rows
# By default, it is the 1st 5 rows, but you can see how many ever rows you want by specifying a number as an agrument

df.head()
```

```python
# Preview last rows

df.tail()
```

```python
# To get random values

df.sample(10)
```

```python
# To get the index of all rows

df.index.tolist()
```

```python
# To get some useful metrics for the data

df.describe()
```

```python
# To find number of unique values

df.nunique()
```

```python
# To find unique values from a column
df['Outcome'].unique()
```

```python
# To get shape of dataframe

df.shape
```

```python
df.info()
```

```python
# To get specific rows and columns
# Python Slices also work

df.loc[5:10, ["Pregnancies", "BloodPressure", "Insulin"]]
```

```python
# iloc is same as loc, but uses indices for columns instead

df.iloc[5:10, 2:6]
```

```python
# To change a specific value

df.loc[1, "Pregnancies"] = 0

df.head()
```

```python
# To sort values

df.sort_values(["Pregnancies", "Glucose"], ascending=False)
```

```python
# To iterate through rows

for index, row in df.iterrows():
    print(index)
    print(row)
    print("\n")
```

```python
df[(df['Pregnancies'] < 5) & (df['Insulin'] > 0)].sort_values(["Pregnancies"], ascending = False)
```

You can do df["column"].str.contains("regex") to find string values in the column with particular patterns.
You can also put case=False and regex=False as additional parameters respectively.

Similar for .str.startsWith("string")

```python
# TO drop a column

df.drop(columns=["BloodPressure"])

# This doesnt actually change it, to make the change, put inplace=True
```
