# Matplotlib Demo

- Data visualization package for python.
- Used for plotting, both 2D and 3D as well as images.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style

from mpl_toolkits.mplot3d import Axes3D
%matplotlib widget
```

```python
# using different stylesheets

style.use('ggplot')
```

```python
# Creating a dataset
x_data = np.random.random(100) * 100
y_data = np.random.random(100) * 100

x_data
y_data
```

```python
# Demo for a scatter plot
# x vs y, color, marker, size, alpha(transparency)
fig1 = plt.figure()
ax1 = fig1.add_subplot(1, 1, 1)
ax1.scatter(x_data, y_data, c='#000000', marker='*', s=100, alpha=0.3)
```

```python
# Demo for a line chart

years = [2010 + x for x in range(16)]
weights = [80, 83, 84, 86, 82, 79, 81, 88, 90, 91, 82, 87, 88, 89, 88, 81]
weight_ticks = list(range(75, 100, 2))

# x vs y, color, line width, line style
fig2 = plt.figure()
ax2 = fig2.add_subplot(1, 1, 1)
ax2.plot(years, weights, c='r', lw=3, linestyle='--', label='Person 1')
ax2.set_title('weight of Mr. X(in KG)', fontsize=30, fontname='Arial')
ax2.set_xlabel('year')
ax2.set_ylabel('Weight(kg)')
ax2.set_yticks(weight_ticks, [f"{x}k USD" for x in weight_ticks])
ax2.legend(loc = 'upper center')
```

```python
# Demo for a bar chart
# Makes sense for categorical data

x = ['C++', 'Java', 'Python', 'Go', 'C']
y = [20, 40, 10, 5, 7]

# x vs y, color, alignment of bar with text, bar width, edge color of line, line width
fig3 = plt.figure()
ax3 = fig3.add_subplot(1, 1, 1)
ax3.bar(x, y, color='green', align='edge', width=0.7, edgecolor='black', lw=10)
```

```python
# Demo for a histogram

ages = np.random.normal(20, 1.5, 1000)

#plt.hist(ages, bins=10)
#plt.hist(ages, bins=[ages.min(), 18, 20, 22, ages.max()])
fig4 = plt.figure()
ax4 = fig4.add_subplot(1, 1, 1)
ax4.hist(ages, bins=10, cumulative=True)
```

```python
# Demo for a pie chart

x = ['C++', 'Java', 'Python', 'Go', 'C']
y = [20, 40, 10, 5, 7]
explodes = [0.1, 0, 0, 0, 0]

fig5 = plt.figure()
ax5 = fig5.add_subplot(1, 1, 1)
plt.pie(y, labels=x, explode=explodes, autopct='%.2f', pctdistance=1.3, startangle=90)
```

```python
# Demo for a box plot

heights = np.random.normal(172, 8, 300)

fig6 = plt.figure()
ax6 = fig6.add_subplot(1, 1, 1)
plt.boxplot(heights)
```

```python
# Demo for subplots

fig = plt.figure()
axs1 = fig.add_subplot(3, 2, 1)
axs2 = fig.add_subplot(3, 2, 2)
axs3 = fig.add_subplot(3, 2, 3)
axs4 = fig.add_subplot(3, 2, 4)
axs5 = fig.add_subplot(3, 2, 5)
axs6 = fig.add_subplot(3, 2, 6)

axs1.scatter(x_data, y_data, c='#000000', marker='*', s=100, alpha=0.3)
axs2.plot(years, weights, c='r', lw=3, linestyle='--', label='Person 1')
axs3.bar(x, y, color='green', align='edge', width=0.7, edgecolor='black', lw=2)
axs3.set_xlabel("demo")
axs3.set_title("Bar")
axs4.hist(ages, bins=10, cumulative=True)
axs5.pie(y, labels=x, explode=explodes, autopct='%.2f', pctdistance=1.3, startangle=90)
axs6.boxplot(heights)
fig.suptitle("demo subplotting")
plt.tight_layout() # So that they dont overlap
```

```python
# Demo for a 3d scatter plot

fig7 = plt.figure()
ax7 = fig7.add_subplot(1, 1, 1, projection='3d')
x_d = np.random.random(100) * 100
y_d = np.random.random(100) * 100
z_d = np.random.random(100) * 100

ax7.scatter(x_d, y_d, z_d)
```

```python
# demo for a 3d mesh plot

a = np.arange(-5, 5, 0.1)
b = np.arange(-5, 5, 0.1)

A, B = np.meshgrid(a, b)
C = np.sin(A) * np.cos(B);

fig8 = plt.figure()
ax8 = fig8.add_subplot(1, 1, 1, projection='3d')

ax8.plot_surface(A, B, C, cmap='Spectral')
```

## Images

- Matplotlib can also show imagaes.
- It first reads an image and converts that image into a numpy array. Each pixel will have 4 values, R, G, B, A where R, G, B are the color values converted to a float32 dtype ranging from 0 to 1. A is the alpha or the transparency value of that pixel. It ranges from 0 to 1 where 0 is full transparency and 11 is no transparency. It can also just be RGB.
- Matplotlib suppoorts greyscale images as well in the float32 format
- It is recommended to have the images in png format rather than jpg/jpeg format.
- You can use various pseudocolor schemes as well. This is extremely helpful when going through greyscale or 1 bit images.

```python
# Demo to show how to import an image and convert it into a np array

import matplotlib.pyplot as plt
import matplotlib.image as mpimg

img = mpimg.imread('C:/Work/ml-research-project/Matplotlib_demo_img.png')

img
```

```python
# Demo to take the np array and plot it as an image

fig9 = plt.figure()
ax = fig9.add_subplot(111)
style.use('default')
ax.imshow(img)
```

```python
from PIL import Image

fig10 = plt.figure()
ax = fig10.add_subplot(111)
style.use('default')
img = Image.open('C:/Work/ml-research-project/Matplotlib_demo_img.png')
img.thumbnail((64, 64), Image.LANCZOS)
ax.imshow(img)

fig11 = plt.figure()
ax11 = fig11.add_subplot(111)
ax11.imshow(img, interpolation='bicubic') # makes it blurry

fig12 = plt.figure()
ax12 = fig12.add_subplot(111)
ax12.imshow(img, interpolation='nearest') # takes the nearest values to add information to missing parts.
```
